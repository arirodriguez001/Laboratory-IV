```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator, FormatStrFormatter

from scipy.signal import find_peaks
```

```python
df = pd.read_csv('', delimiter=',', header=None)
df = df.apply(pd.to_numeric, errors="coerce")  # Convierte todo a números y reemplaza errores con NaN

tiempo = np.asarray(df[1], dtype=float)
vin = np.asarray(df[2], dtype=float)
vout = np.asarray(df[3], dtype=float)

plt.plot(tiempo, vin, label="Vin")
plt.plot(tiempo, vout, label="Vout")

plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
"""
Si la señal que estoy mandano es un Sen(x), al integrar esperaria recibir un -Cos(x) pero el integrador cambia el signo por lo que me quedo con el cos(x).
"""

plt.plot(tiempo, vin, label="Vin")
plt.plot(tiempo, -vout + 9, label="Vout")

plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Para hallar los picos

plt.plot(tiempo, vin, label="Vin")

peaks, properties = find_peaks(vin, height=2) #cambiar
print("Posiciones de los máximos (en x):", tiempo[peaks])

plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.plot(tiempo[peaks], vin[peaks], "ro", label="Máximos")

plt.legend()
plt.grid(True)

plt.show()
```

```python
# Tambien puede funcinonar

plt.plot(tiempo, vout, label="Vout")

peaks, properties = find_peaks(vout, height=11.1)

filtered_peaks = [p for p in peaks if -0.04 < tiempo[p] < -0.02]

print("Posiciones de los máximos (en x, filtrados):", tiempo[filtered_peaks])

plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")

plt.plot(tiempo[filtered_peaks], vout[filtered_peaks], "ro", label="Máximos")

plt.legend()
plt.grid(True)

plt.show()
```
