```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.fft import fft, fftfreq
import pandas as pd
```

```python
file_path = "C:/Users/FIUBA/Desktop/Labo 4/mod young/mediciones24-2/med7-24-2-cobre-22-7cm(mala creo)0.csv"
df = pd.read_csv(file_path)

df.columns = ["t","V"]
```

```python
plt.plot(df["t"],df["V"])
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df["t"], df["V"], '.', markersize=2, color="orange")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [V]")
plt.title("Datos originales")
plt.show()

# Seleccionar los ltimos puntos para el ajuste lineal (por ejemplo, ltimos 20% de los datos)
fraction = 0.2
n_points = int(len(df) * fraction)
df_tail = df.iloc[-n_points:]

# Ajuste lineal (mnimos cuadrados)
p = np.polyfit(df_tail["t"], df_tail["V"], 1)  # p[0] es la pendiente, p[1] es la interseccin

# Crear la recta ajustada
recta_ajustada = np.polyval(p, df["t"])

# Restar la recta ajustada a los datos originales
df["V1_corregido"] = df["V"] - recta_ajustada

plt.figure(figsize=(6, 4))
plt.plot(df["t"], df["V1_corregido"], '.', markersize=2, color="blue")
plt.axhline(np.mean(df["V1_corregido"]), color='red', linestyle='--', label="Media")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje corregido [V]")
plt.title("Datos despus de restar la tendencia lineal")
plt.show()

# Mostrar la ecuacin de la recta ajustada
print(f"Ecuacin de la recta ajustada: V(t) = {p[0]:.6f} * t + {p[1]:.6f}")
print(f"mean: {np.mean(df['V1_corregido'])}")
```
